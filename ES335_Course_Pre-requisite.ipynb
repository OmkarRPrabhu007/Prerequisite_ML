Question 2 :

```python
def matrixMultiplication(A, B):
    result = [[0 for j in range(len(B[0]))] for i in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]

    return result

A = [[1, 2, 3,1,2,3], [4, 5, 6,1,2,3], [7, 8, 9,1,2,3],[1, 2, 3,1,2,3], [4, 5, 6,1,2,3], [7, 8, 9,1,2,3]]
B = [[10, 11, 12,1,2,3], [13, 14, 15,1,2,3], [16, 17, 18,1,2,3],[10, 11, 12,1,2,3], [13, 14, 15,1,2,3], [16, 17, 18,1,2,3]]

start_time = time.time()

result = matrixMultiplication(A, B)

end_time = time.time()

for row in result:
    print(row)


print(f"Time taken for the execution of matrix multiplication by normal list method is {end_time - start_time}")
```

```python
a = np.array(A)
b = np.array(B)

start_time = time.time()

numpy_result = a.dot(b)

end_time = time.time()

for row in numpy_result:
    print(row)

print(f"Time taken for the execution of matrix multiplication by numpy method is {end_time - start_time}")
```

Numpy are faster considered to that of normal mrthod as Normal method multiplies element wise and it takes too much time. Whereas, the numpy operations are vectorised and it works on the entire array saving its time.

```python
def findMedian1(arr) :
    n = len(arr)
    
    for i in range(len(arr)//2 -1) :
        max = arr[0] 
        for i in range(len(arr)) :
            if(arr[i] > max) :
                max = arr[i]
        arr.remove(max)

    max1 = arr[0]
    for i in range(len(arr)) :
        if(arr[i] > max1) :
            max1 = arr[i]

    arr.remove(max1)
    max2 = 0
    for i in range(len(arr)) :
        if(arr[i] > max2) :
            max2 = arr[i]

    median = (max1 + max2)/2

    if(n%2 != 0) :
        median = max2 
    else :
        median = (max1 +max2)/2
    
    return median

np_arr = range(1,1000)
arr = list(np_arr)

start_time = time.time()

median = findMedian1(arr)

end_time = time.time() 

print(f"The Median is {median} by one pass method.")    
print(f"Time taken by the above method is {end_time - start_time}")
```

Here, the time complexity is O(n^2)

```python
def findMedian2(arr):
    arr.sort()
    n = len(arr)
    
    if n % 2 == 0:
        median = (arr[n//2 - 1] + arr[n//2]) / 2
    else:
        median = arr[n//2]
        
    return median

np_arr = range(1,1000)
arr = list(np_arr)


start_time = time.time()

median = findMedian2(arr)

end_time = time.time()

print(f"The Median is {median} by sorting method.")  
print(f"Time taken by the above method is {end_time - start_time}")
```

Here, the time complexity is O(nlog(n))

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, name, roll_number, marks):
        self.students.append(Student(name, roll_number, marks))

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, name):
        self.branches.append(Branch(name))

class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(Year(year))
```
